```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist
```

```python
text = "Language modeling is a important part of N atural Language Processing"
```

```python
tokens = word_tokenize(text.lower())
```

```python
bigrams = list(ngrams(tokens, 2))
```

```python
freq = FreqDist(bigrams)
```

```python
print("Bigrams and their frequencies:")
for k, v in freq.items():
  print(k, ":", v)
```

```python
print("\nMost Common Bigrams:")
print(freq.most_common(5))
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist
```

```python
text = "Natural Language processing enables machines to understand and generate human Language"
```

```python
tokens = word_tokenize(text.lower())
```

```python
unigrams = tokens
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))
```

```python
print("unigrams:", unigrams)
print("bigrams:", bigrams)
print("trigrams:", trigrams)
```

```python
print("\nMost common unigrams:")
print(FreqDist(unigrams).most_common(5))
```

```python
print("Most Common Bigrams:")
print(FreqDist(bigrams).most_common(5))
```

```python
print ("Most Common Trigrams:")
print(FreqDist(trigrams).most_common(5))
```
