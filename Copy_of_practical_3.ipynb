```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
text="Natural Lamguage Processing enables machines to understand human languages."
```

```python
tokens = word_tokenize(text)
```

```python
freq_dist = FreqDist(tokens)
```

```python
print("Word Frequencies:", freq_dist)
print("Most Common Words:", freq_dist.most_common(5))
print("Vocabulay_size", len(freq_dist))
```
